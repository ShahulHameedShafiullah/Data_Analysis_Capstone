```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT *
FROM DevType
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df['DevType'].value_counts()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df_table = pd.read_sql_query(QUERY,conn)
df_table
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
data = pd.read_sql_query(QUERY,conn)
data.info()
```

```python
# Plot a histogram for ConvertedComp
plt.hist(data['ConvertedComp'], histtype="bar", align="mid", orientation="vertical", rwidth=None, color="r", bins=10) 
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT * FROM master
"""
data = pd.read_sql_query(QUERY,conn)
data.boxplot(column='Age', vert=False)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

import seaborn as sns

QUERY = """
SELECT * FROM master
"""
data = pd.read_sql_query(QUERY,conn)
plot = sns.scatterplot(x='Age', y='WorkWeekHrs', data=data)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""
data=pd.read_sql_query(QUERY,conn)

sns.scatterplot(data=data, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=0.5, sizes=(30, 200))
plt.title('Work Week Hrs and Code Rev Hrs by Age', size=12) 
plt.xlabel('WorkWeekHrs', size=10) 
plt.ylabel('CodeRevHrs', size=10)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """ 
SELECT DatabaseDesireNextYear, COUNT(*) as count 
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5
""" 

data = pd.read_sql_query(QUERY,conn)  
data.set_index('DatabaseDesireNextYear', inplace=True) 

colors_list=['lightgreen', 'yellow', 'coral', 'skyblue', 'red', 'gold']

data['count'].plot(kind='pie', figsize=(20,10), autopct='%1.1f%%', labels=None, startangle=0, colors=colors_list, shadow=True, pctdistance=1.12)  

plt.legend(labels=data.index, loc='upper right')
plt.title('Top 5 Databases Respondents Wish To Learn Next Year') 
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT Age, CodeRevHrs, WorkWeekHrs FROM master
WHERE Age BETWEEN 30 AND 35
"""
data = pd.read_sql_query(QUERY,conn)
df_sc = data.groupby('Age').median()

df_sc.plot(kind='bar', figsize=(8, 4), stacked=True)

plt.title('Median of Work Week Hrs and Code Rev Hrs for Age Group 30 to 35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT Age, ConvertedComp FROM master
WHERE Age BETWEEN 45 AND 60
"""
data = pd.read_sql_query(QUERY,conn)
df_LC = data.groupby('Age').median()

df_LC.plot(kind='line', figsize=(8, 4), marker='o', linestyle='-')

plt.title('Median of Converted Compensation for Age Group 45 to 60')
plt.xlabel('Age')
plt.ylabel('Converted Compensation')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
QUERY = """
SELECT Age, ConvertedComp FROM master
WHERE Age BETWEEN 25 AND 30
"""
data = pd.read_sql_query(QUERY,conn)
df_LC = data.groupby('Age').median()

df_LC.plot(kind='line', figsize=(8, 4), marker='o', linestyle='-')

plt.title('Median of Converted Compensation for Age Group 45 to 60')
plt.xlabel('Age')
plt.ylabel('Converted Compensation')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
